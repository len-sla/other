````python
# and funcion doing removing extra spaces and new lines

def remove_extra(temp):
  temp = re.sub(r'\n+', ' ', temp) #removing extra new lines
  temp = re.sub(r' +', ' ', temp)  #removing extra spaces
  return temp
````
